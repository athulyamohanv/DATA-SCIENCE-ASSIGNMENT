```python
Q1=data['LB'].quantile(0.25)
Q3=data['LB'].quantile(0.75)
IQR=Q3-Q1
print(Q1,Q3,IQR)
```

```python
lower_extreme=Q1-1.5*IQR
upper_extreme=Q3+1.5*IQR
print(lower_extreme,upper_extreme)
```

```python
data_lb_outliers=data[(data['LB']<lower_extreme) | (data['LB']>upper_extreme)]
data_lb_outliers
```

```python
data_no_outliers=data[(data['LB']>=lower_extreme) & (data['LB']<=upper_extreme)]
data_no_outliers
data_no_outliers.boxplot('LB')
plt.show()
```
